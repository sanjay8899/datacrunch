```
df=pd.read_csv(r"/content/gdrive/My Drive/data crunch/train.csv")
```

```
df.head
df.fillna(df.mean())
```

```
df=df.drop('Loan.ID',axis=1)
#df=df.dropna()
```

```
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder();
#df1.apply(LabelEncoder().fit_transform)
#label.fit_transform(df1)
#df1['city']=label.fit_transform(df1['city'])
cols=[];
for col in df.columns.values:
  if df[col].dtype not in ['int64','float64']:
  #  df1.head()
    cols.append(col)
    #df1[col]=label.fit_transform(df1[col])
for col in cols:
  #print(col)
  df[col]=label.fit_transform(df[col].astype(str))
```

```
df=df.fillna(df.mean())
df.head()
#df.shape
```

```
y=df['Loan.Status']
x=df.drop('Loan.Status',axis=1)
x.head()
x.shape
print(y)
```

```
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```
print(x)
```

```
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=32,activation='relu',input_shape=(16, )))
model.add(tf.keras.layers.Dense(units=1,activation='softmax'))
model.compile(optimizer='adam',loss="MeanSquaredError")
model.summary()
```

```
model.fit(x,y,epochs=5)
```
